```python
# Clustering (Univariate, bivariate and multivariate)
```

```python
clustering1 = KMeans(n_clusters = 6) #Initial
clustering1.fit(df[['Annual Income (k$)']]) # Fit
# Grab labels
```

```python
df['Income clusters'] = clustering1.labels_
```

```python
df.head()
```

```python
df['Income clusters'].value_counts()
```

```python
clustering1.inertia_ #distance between the centroids. Elbow method.
```

```python
inertia_score = []
for clust in range(1, 11):
    kmeans = KMeans(n_clusters = clust)
    kmeans.fit(df[['Annual Income (k$)']])
    inertia_score.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), inertia_score)
```

```python
clustering = KMeans(n_clusters = 3) #Initial
clustering.fit(df[['Annual Income (k$)']]) # Fit
df['Income clusters'] = clustering.labels_
df['Income clusters'].value_counts()
```

```python
clustering2 = KMeans() #Initial
clustering2.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']]) # Fit
df['Spending & Income clusters'] = clustering2.labels_
df['Spending & Income clusters'].value_counts()
```

```python
inertia_score2 = []
for clust in range(1, 11):
    kmeans = KMeans(n_clusters = clust)
    kmeans.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
    inertia_score2.append(kmeans.inertia_)
plt.plot(range(1, 11), inertia_score2)
```

```python
clustering2 = KMeans(n_clusters=5) #Initial
clustering2.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']]) # Fit
df['Spending & Income clusters'] = clustering2.labels_
df['Spending & Income clusters'].value_counts()
```

```python
centers = pd.DataFrame(clustering2.cluster_centers_)
plt.scatter(x = centers[0], y = centers[1], s = 100, marker ='+' , c = 'black')
sb.scatterplot(data = df, x = 'Annual Income (k$)', y = 'Spending Score (1-100)', hue = 'Spending & Income clusters', palette = 'tab10')

```

```python
# normalised by index to get percentages
pd.crosstab(df['Spending & Income clusters'], df['Gender'], normalize='index')
```

```python
# Multivariate Clustering
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
dff = pd.get_dummies(df,drop_first=True)
dff.head()
```

```python
dff.columns
```

```python
dff = dff[['Age', 'Annual Income (k$)', 'Spending Score (1-100)','Gender_Male']]
```

```python
dff = pd.DataFrame(scale.fit_transform(dff))
```

```python
dff.head()
```

```python
inertia_score3 = []
for clust in range(1, 11):
    kmeans = KMeans(n_clusters = clust)
    kmeans.fit(dff)
    inertia_score3.append(kmeans.inertia_)
plt.plot(range(1, 11), inertia_score3)
```
